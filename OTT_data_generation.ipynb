```python

import pandas as pd
import random
from faker import Faker
from random import randrange
from datetime import datetime

nr_of_customers = 40000

fake = Faker('de_DE')

customers = []

for customers_id in range(nr_of_customers):

    # Create User_ID
    user_ID = fake.ean(length=8)
    #create user's's name
    name = fake.name()
    # Create gender
    gender = random.choice(["M", "F"])
    
    # Create date of birth 
    d1 = datetime.strptime(f'1/1/1947', '%m/%d/%Y')
    d2 = datetime.strptime(f'8/10/2003', '%m/%d/%Y')
    date_of_birth = fake.date_between(d1, d2)

    # Create email 
    email = fake.ascii_email()

    #Create city
    city = fake.city()

    #Create state
    state = fake.state()
    #Create country
    country = fake.country()

    # Create Signupdate 
    d1 = datetime.strptime(f'1/1/2007', '%m/%d/%Y')
    d2 = datetime.strptime(f'8/10/2022', '%m/%d/%Y')
    signup_date = fake.date_between(d1, d2) 
    
    # Create Transactiondate
    d1 = datetime.strptime(f'1/1/2007', '%m/%d/%Y')
    d2 = datetime.strptime(f'8/10/2022', '%m/%d/%Y')
    Transaction_date = fake.date_between(d1, d2)

    # Create cancel date 
    d1 = datetime.strptime(f'1/1/2008', '%m/%d/%Y')
    d2 = datetime.strptime(f'8/10/2022', '%m/%d/%Y')
    cancel_date = fake.date_between(d1, d2)

    # Create status
    status = random.choice(["Active", "Inactive"])

    #create amount spent
    amount_spent = fake.pyfloat(right_digits=2, positive=True, min_value=1, max_value=100)

    #create amount gained
    amount_gained = fake.pyfloat(right_digits=2, positive=True, min_value=1, max_value=1000)


#     my_word_list = [
#     'danish','cheesecake','sugar',
#     'Lollipop','wafer','Gummies',
#     'sesame','Jelly','beans',
#     'pie','bar','Ice','oat' ]

# fake.sentence()
# # 'Expedita at beatae voluptatibus nulla omnis.'

# fake.sentence(ext_word_list=my_word_list) 

    customers.append([user_ID , name, gender, date_of_birth, email, city,state,country,signup_date, Transaction_date,cancel_date, status, amount_spent,amount_gained])

customers_df = pd.DataFrame(customers, columns=['user_ID' , 'name', 'gender', 'date_of_birth', 'email', 'city','state','country','signup_date', 'Transaction_date','cancel_date', 'status', 'amount_spent','amount_gained']) 
                
pd.pandas.set_option('display.max_columns', None)
# print(customers_df)
```

```python

import pandas as pd
import random
from faker import Faker
from random import randrange
from datetime import datetime

nr_of_OTT = 10000000

fake = Faker('de_DE')

OTT = []

for i in range(nr_of_OTT):

    # Create User_ID
    user_ID = random.choice(customers_df.user_ID)
    # date 
    d1 = datetime.strptime(f'1/1/1947', '%m/%d/%Y')
    d2 = datetime.strptime(f'8/10/2003', '%m/%d/%Y')
    date = fake.date_between(d1, d2)

    # Create engagement 
    engagement = random.choice(['Yes','No'])

    #Program watched
    Program = random.choice(netflix.title)

    #STart time
    start_time=fake.time()

    #end time
    end_time=fake.time()

    #Create Device viewership
    Device_viewership = random.choice(['mobile', 'ROKU',' Apple TV'])

    OTT.append([user_ID ,date, engagement,Program,start_time,end_time,Device_viewership])

OTT = pd.DataFrame(OTT, columns=['user_ID' ,'date', 'engagement','Program','start_time','end_time','Device_viewership']) 
                
pd.pandas.set_option('display.max_columns', None)
# print(customers_df)
```
